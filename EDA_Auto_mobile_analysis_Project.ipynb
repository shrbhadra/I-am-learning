## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
import pandas as pd
# Check for missing values
missing_values = dataset.isnull().sum()

# If there are missing values, handle them
if missing_values.any():
  # Drop rows with missing values
  dataset = dataset.dropna()

# Check for duplicate values
duplicate_rows = dataset[dataset.duplicated()]

# If there are duplicate values, remove them
if duplicate_rows.shape[0] > 0:
  dataset = dataset.drop_duplicates()



# Standardize the numerical variables
numerical_variables = dataset.select_dtypes(include=['int64', 'float64'])
numerical_variables = (numerical_variables - numerical_variables.mean()) / numerical_variables.std()

# Combine the categorical and numerical variables
dataset = pd.concat([numerical_variables, dataset.select_dtypes(include=['object'])], axis=1)

# Print the first few rows of the cleaned dataset
dataset.head()
````

````python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.colors import rgb2hex

# Load the dataset
dataset = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/automobile_data.csv')

# Create a list of colors
colors = list(cm.tab10.colors)
colors = [rgb2hex(color) for color in colors]

# Create a dictionary of car makes and their counts
car_makes = dataset['make'].value_counts().to_dict()

# Create a pie chart of car makes
plt.figure(figsize=(10, 7), dpi=100)
plt.pie(list(car_makes.values()), labels=list(car_makes.keys()), autopct='%1.1f%%', colors=colors)
plt.title('Car Makes Distribution', fontsize=16)
plt.axis('equal')  # Equal aspect ratio ensures a circular pie chart
plt.axis('off')  # Hides the axis labels

# Add 3D effect
ax = plt.gca()
ax.dist = 11.7  # Adjust this value to control the 3D effect
ax.azim = -75  # Adjust this value to rotate the pie chart

plt.show()
````

````python
# Chart - 8 visualization code
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
dataset = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/automobile_data.csv')

# Select the numerical columns
numerical_columns = dataset.select_dtypes(include=['int64', 'float64',])

# Calculate the correlation matrix
correlation_matrix = numerical_columns.corr()

# Create a heatmap of the correlation matrix
plt.figure(figsize=(10, 7))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Automobile Data')
plt.show()
````

##### 1. Why did you pick the specific chart?

The Correlation Heatmap chart was chosen to visualize the correlations between the numerical columns in the dataset. This allows for easy identification of any potential relationships or patterns between the different variables. Additionally, the Correlation Heatmap chart can be used to identify any variables that are highly correlated or uncorrelated with each other.

##### 2. What is/are the insight(s) found from the chart?

**Insights:**

* There is a weak negative correlation between engine size and city mpg. This means that as engine size increases, city mpg tends to decrease.
* There is a weak negative correlation between engine size and highway mpg. This means that as engine size increases, highway mpg tends to decrease.

* There is a weak positive correlation between city mpg and highway mpg. This means that as city mpg increases, highway mpg also tends to increase.

**Additional Insights:**

* The Correlation Heatmap chart can be used to identify which variables are most strongly correlated with each other.
* The Correlation Heatmap chart can be used to identify which variables are least correlated with each other.
* The Correlation Heatmap chart can be used to identify any potential multicollinearity issues in the data.

**Overall:**

The Correlation Heatmap chart provides a visual representation of the correlations between the numerical variables in the dataset. This information can be used to identify trends and patterns that may not be apparent from looking at the data in a tabular format.
